```python
import urllib.request
import pandas as pd
import json
```

```python
df_bibliothek = json.load(open("Bibliothek.json"))
df_medizin = json.load(open("Medizin.json"))
df_musik = json.load(open("Musik.json"))
df_cat = json.load(open("cat.json"))
df_kunst = json.load(open("kunst.json"))
df_wool = json.load(open("wool.json"))


# nicht über merge sonder über die Variante, die in den Videotutorial vorgestellt 
# wurde unter "Daten zusammenführen"
```

```python
df_bibliothek_selection = pd.DataFrame(df_bibliothek["items"])
df_medizin_selection = pd.DataFrame(df_medizin["items"])
df_musik_selection = pd.DataFrame(df_musik["items"])
df_cat_selection = pd.DataFrame(df_cat["items"])
df_kunst_selection = pd.DataFrame(df_kunst["items"])
df_wool_selection = pd.DataFrame(df_wool["items"])


# Erstellen der DataFrames aus den Listen der Dictionaries, die die Dateneinträge der Bibsonomy Suche darstellen
```

```python
df_bibliothek_selection = df_bibliothek_selection.rename(columns={"label": "title_bib"})
```

```python
df_bibliothek_selection["key"] = "bibliothek"
```

```python
df_medizin_selection["key"] = "medizin"
```

```python
df_cat_selection["key"] = "cat"
```

```python
df_kunst_selection["key"] = "kunst"
```

```python
df_wool_selection["key"] = "Wolle"
```

```python
df_musik_selection["key"] = "musik"

# in den letzten Befehlen wurden die Suchtermini hinzugefügt
```

```python
df_bib = df_bibliothek_selection
df_med = df_medizin_selection
df_mus = df_musik_selection
df_cat = df_cat_selection
df_art = df_kunst_selection
df_woo = df_wool_selection
```

```python
data = pd.concat([df_bib, df_med, df_mus, df_cat, df_art, df_woo], ignore_index=True)
data
```

```python
data.columns
```

# Erstellung des Exportformats

```python
df_data = data[["id", "tags", "label", "description", "author", "publisher", "date", "year", "type", "key"]]
```

```python
df_data
```

```python
df_data = df_data.loc[df_data["author"].isnull() == False]
df_data = df_data.loc[df_data["label"].isnull() == False]

# Daten auswählen, in denen ein Wert (ausgewählte Spalte) eingetragen ist.
```

```python
df_data
```

```python
df_data.loc[df_data.type == "Publication"]

# Ausgabe aller Spalten die type = Publication sind
```

```python
df_data = df_data.rename(columns={"label": "title", "author": "authors"})
df_data
```

```python
df_data = df_data.drop_duplicates(subset ="title", keep="first")
df_data

# Duplikate entfernt
```

```python
df_data["year"].unique()
```

```python
df_data["year_correct"] = df_data.apply(lambda row: row.year.isdigit(), axis=1)
```

```python
df_data.loc[df_data.year_correct ==False]

# Beide Befehle dienen dazu, die Werte die keine Jahreszahlen enthalten zu identifizieren
```

```python
df_data = df_data.loc[df_data.year_correct == True]
df_data
```

```python
df_data["year_number"] = df_data["year"].astype(int)
```

```python
df_data.info()
```

```python
df_data = df_data.loc[df_data.year_number <=2000]
```

```python
df_data
```

```python
df_data = df_data.dropna()
```

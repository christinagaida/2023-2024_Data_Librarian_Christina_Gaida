### Folgendes wird benötigt für die idf

```python
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
```

```python
stem = stemmer.stem("bildanalyse")
stem

# Grundform (laut dem System) bilden
```

```python
# In wie vielen Dokumenten kommen diese vor?

lexicon = index.getLexicon()
lexicon[stem].getDocumentFrequency()
```

```python
# Abfrage wie viele Dokumente im Index sind (für idf)
# Kann man mit der oberen Statistic überprüfen

index.getCollectionStatistics().numberOfDocuments
```

```python
def get_idf_for_term(term, index):
    lex = index.getLexicon()
    stemmed_term = stemmer.stem(term)

    if not stemmed_term in lex:
        return

    lex_entry = lex[stemmed_term] # Lexikon-Eintrag

    df_term = lex_entry.getDocumentFrequency() # darauf anwenden
    N = index.getCollectionStatistics().numberOfDocuments

    #inverse df
    idf = N/df_term

    # Log10 anwenden
    log_idf = np.emath.logn(10, idf)

    return log_idf
```

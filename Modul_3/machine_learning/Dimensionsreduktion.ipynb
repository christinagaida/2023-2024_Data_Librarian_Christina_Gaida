```python
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
from sklearn.decomposition import PCA
pca = PCA(random_state=1, n_components=2)

# Methode PCA (Principle Component Analysis) zur Dimensionsrediktion
# Nur 2 Dimensionen/Components sollen erzeugt werden
```

```python
pca_result = pca.fit_transform(digits.data) # Transformation wird durchgef√ºhrt
```

```python
pca_result.shape # Matrix mit der gleichen Anzahl von Datenpunkten, aber nur 2 Spalten
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
fig, ax = plt.subplots()
plt.scatter(pca_result[:, 0], pca_result[:, 1], c=digits.target, cmap="Set1")
plt.colorbar()

# Spalten der resultierenden Matrix werden geplottet
```

```python
from sklearn.manifold import TSNE
tsne = TSNE(random_state=1, n_components=2)
tsne_result = tsne.fit_transform(digits.data)

# gleiches Vorgehen mit dem t-SNE-Verfahren
```

```python
fig, ax = plt.subplots()
plt.scatter(tsne_result[:, 0], tsne_result[:, 1], c=digits.target, cmap="Set1")
plt.colorbar()

# Ergebnis graphisch darstellen
```
